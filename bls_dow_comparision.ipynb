```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dow_data = pd.read_csv("outputs/clean_dow.csv")
bls_1996_2016 = pd.read_csv("outputs/bls_1996_2016.csv")
dow_1996_2016 = dow_data.loc[dow_data['Year']>1995]
dow_1996_2016
```

```python
bls_1996_2016
```

```python
dow_bar = plt.bar(x =dow_1996_2016['Year'], height=dow_1996_2016['Rate Change'], 
                   color=(dow_1996_2016['Rate Change'] > 0).map({True: 'c',False: 'r'}))

plt.rcParams["figure.figsize"] = (20, 12)
plt.title("Dow Stock - Rate Change(%) 1996-2016")
plt.xlabel("Years")
plt.ylabel("Dow - Yearly Rate Change(%)")
plt.xticks(dow_1996_2016['Year'])
plt.grid()

plt.savefig('outputs/dow_bar.png')
```

```python
bls_bar = plt.bar(bls_1996_2016['year'], bls_1996_2016['Rate Change'],
                   color=(bls_1996_2016['Rate Change'] > 0).map({True: 'c',False: 'r'}))
plt.rcParams["figure.figsize"] = (20, 12)
plt.title("Unemployment - Rate Change(%) 1996-2016")
plt.xlabel("Years")
plt.ylabel("Unemployment - Yearly Rate Change(%)")
plt.xticks(bls_1996_2016['year'])
plt.grid()
plt.savefig('outputs/Unemployment_bar.png')
```
