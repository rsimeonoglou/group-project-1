```python
responses_20162 = responses_2016[0]['data']

bls_df_3 = pd.DataFrame(responses_20162)
bls_df_3
```

```python
dataframes = [bls_df_3,bls_df_2,bls_df]
bls_df_final = pd.concat(dataframes)
bls_df_final
```

```python
bls_df_by_year = bls_df_final.loc[(bls_df_final['period'] == 'M12')|
                                  (bls_df_final['period'] == 'M01')
                                 ].reset_index(drop=True)

bls_df_by_year
```

```python
bls_df_by_year['Year Change'] = ""

bls_df_by_year['value'] = bls_df_by_year['value'].astype(float)

# use iterrows to iterate through pandas dataframe
start_value = 0

for index, row in bls_df_by_year.iterrows():
    
    change =  start_value - row['value'] 
    start_value = row['value']
    bls_df_by_year.loc[index, 'Year Change'] = change
    
bls_df_by_year
    
```

```python
bls_df_change = bls_df_by_year.loc[(bls_df_by_year['period'] == 'M01')].reset_index(drop=True)
bls_df_change = bls_df_change[['year','Year Change']] 
bls_df_change
```
